```python
import numpy as np
import matplotlib.pyplot as plt
G=1
M=1
x=1
y=0
vx=0
vy=1
t=0
dt=0.01
x_list=[x]
y_list=[y]
while t<20:
    r=np.sqrt(x**2+y**2)
    ax=-G*M*x/r**3 #cost=x/r
    ay=-G*M*y/r**3
    vx=vx+ax*dt
    vy=vy+ay*dt
    x=x+vx*dt
    y=y+vy*dt
    x_list.append(x)
    y_list.append(y)
    t=t+dt
plt.figure(figsize=(6,6))
plt.plot(x_list,y_list,label="Earth Trajectory")
plt.plot(0,0,"ro",markersize=10,label="Sun")
plt.title("2-Body Simulation: Earth & Sun")
plt.xlabel("X position")
plt.ylabel("Y position")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
