```python
#use Monte Carlo to get pi
import numpy as np
N=100000               #第一次取太大把c盘占满了。。。。。
x=np.random.rand(N)
y=np.random.rand(N)
r=1
inside=np.sqrt(x**2+y**2)<=r
P=(np.sum(inside))/N
pi=4*P                        #because (1/4*pi*r**2)/r**2=p
print('pi='+str(pi))
```

```python
#可视化（visualization）;分开是为了上面N可以取较大的值，用有限的算力尽可能的算准pi
import numpy as np
import matplotlib.pyplot as plt
N=5000          
x=np.random.rand(N)
y=np.random.rand(N)
r=1
inside=np.sqrt(x**2+y**2)<=r
plt.figure(figsize=(5,5))
plt.scatter(x[inside],y[inside],s=3,color='red',label='Inside')
plt.scatter(x[~inside],y[~inside],s=3,color='blue',label='Outside')
plt.title("Visualization")
plt.legend()
plt.axis('equal')
plt.show()
```
